```python
import pybaseball as pyb
```

```python
pitching = pyb.pitching_stats(start_season=2024, qual=120)
```

```python
pitching.filter(items=["ERA", "G", "WHIP", "K%", "BB%", "K-BB%"]).mean()
```

```python
pitching.loc[pitching.ERA > 4.0].filter(items=["Name", "Team", "ERA", "xERA", "WHIP", "GS", "IP", "W", "K%", "BB%", "K-BB%"]).sort_values("xERA", ascending=True).head(10)
```

```python
pitching.filter(items=["Name", "Team", "GS", "IP", "ERA", "xERA", "WHIP", "SO", "K%"]).sort_values("xERA", ascending=True).head(50)
```

```python
recent = pyb.pitching_stats(start_season=2020, end_season=2024, qual=40)
recent.loc[recent.Name == "Tanner Banks"].filter(items=["Season", "IDfg", "Name", "Team", "ERA", "IP", "GS", "SO", "K%"]).sort_values("Season", ascending=False)
```
